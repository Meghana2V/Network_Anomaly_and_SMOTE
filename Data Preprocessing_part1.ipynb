### Converting em Datatypes

```python
import hashlib

def hash_data(data):
    # Create an MD5 hash object
    md5 = hashlib.md5()
    # Update the hash object with the encoded string
    md5.update(data.encode('utf-8'))
    # Get the hexadecimal digest of the hash
    hex_digest = md5.hexdigest()
    # Convert the first 8 characters of the hex digest to an integer
    encoded_int = int(hex_digest[:16], 16)
    return encoded_int

# Apply the hash_data function to the entire column
data_all_new['Flow_ID_Encoded'] = data_all_new['Flow ID'].apply(hash_data)
data_all_new['Source_IP_Encoded'] = data_all_new['Source IP'].apply(hash_data)
data_all_new['Destination_IP_Encoded'] = data_all_new['Destination IP'].apply(hash_data)

print("\n",data_all_new['Flow_ID_Encoded'])
print("\n",data_all_new['Source_IP_Encoded'])
print("\n",data_all_new['Destination_IP_Encoded'])
```

```python
from datetime import datetime

def time_encode(data):
    formats = ['%d/%m/%Y %H:%M:%S', '%d/%m/%Y %H:%M']
    
    for fmt in formats:
        try:
            # Try to convert timestamp string to datetime object
            dtime = datetime.strptime(data, fmt)
            # Convert datetime object to POSIX timestamp in seconds
            dtimestamp = dtime.timestamp()
            return int(round(dtimestamp))
        except ValueError:
            continue
    
    # If no format matches, return None or handle it as needed
    print(f"Error: Timestamp format not recognized for {data}")
    return None


# Apply the time_encode function to the entire column
data_all_new['Timestamp'] = data_all_new['Timestamp'].apply(time_encode)
print(data_all_new['Timestamp'])
```
